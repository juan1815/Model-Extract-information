```python
import pdfplumber
```

```python
def extract_text_and_coordinate(pdf_path):
    text_with_coordinates = []
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            words = page.extract_words()
            for word in words:
                text_with_coordinates.append({
                    "text": word["text"],
                    "x0": word["x0"],
                    "y0": word["top"],
                    "x1": word["x1"],
                    "y1": word["bottom"]
                })
    return text_with_coordinates
```
